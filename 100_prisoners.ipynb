# The 100 Prisoners Riddle
There are 100 prisoners sentenced to life in prison. There’s a central living room with one light bulb which is is initially off. No prisoner can see the light bulb from his or her own cell. Everyday, the warden picks a prisoner equally at random, and that prisoner visits the living room.

While there, the prisoner can toggle the bulb if he or she wishes. The prisoner has the option of asserting that all 100 prisoners have been to the living room by now. If this assertion is false, all 100 prisoners are shot.

If it is indeed true, all prisoners are set free and inducted into MENSA, since the world could always use more smart people. Thus, the assertion should only be made if the prisoner is 100% certain of its validity. The prisoners are allowed to get together one night in the courtyard, to discuss a strategy. What plan should they come up with so that someone will make a correct assertion?

# Solution
* The prisoners need to choose a captain or manager who will be in charge of counting.
* When a prisoner, who is other than the manager, enters the room he or she will turn the light on if the light is off and he has never switched the light on.
* If the light is off when the manager enters the room, he will leave it untouched. If the light is on, the manager will turn it off and add one to his count.
* When the manager’s count gets to 99 he will state that all 100 prisoners have entered the room and they will all be released from prison.

```python
import pandas as pd
import numpy as np
```

```python
class prison:
  def __init__(self):
    self.cells = {}
    self.manager = np.random.randint(100)
    self.day_count = 0
    self.manager_count = 0
    print("Manager is:",self.manager)

  def build_prison(self):
    for i in range(100):
      self.cells[i] = {'first':False}

  def run(self):

    self.manager_count = 0
    self.day_count = 0
    lightswitch = False
    decision = False

    while not decision:
      self.day_count += 1
      prisoner = np.random.randint(100)
      print("Day: ",self.day_count,"\tPrisoner: ",prisoner, end='')
      if prisoner == self.manager and lightswitch == True:
        lightswitch = False
        self.manager_count += 1
        print(" NEW! manager_count: ", self.manager_count)
        if self.manager_count == 99:
          decision = True
      else:
        if lightswitch == False and self.cells[prisoner]['first'] == False:
          lightswitch = True
          self.cells[prisoner]['first'] = True
          print(" Prisoner ", prisoner, "turned on the switch")
        else:
          print("")
    print("All prisoners have entered the room after:",self.day_count)
```

```python
Model = prison()
Model.build_prison()
```

```python
Model.run()
```
